```python
# 阶段案例——反恐精英
class Person():
    def __init__(self,name):
        # 士兵和敌人都有初始血量100
        self.name = name
        self.blood = 100
    # 给弹夹安装子弹
    def install_bullet(self,clip,bullet):
        # 战士安装子弹到子弹夹
        clip.save_bullet(bullet)
    # 给枪安装弹夹
    def install_clip(self,gun,clip):
        gun.mounting_clip(clip)
    def take_gun(self,gun):
        self.gun = gun
        # 士兵拿枪
    def fire(self,enemy):
        #射击敌人
        self.gun.shoot(enemy)
    def __str__(self):
        return self.name + "剩余血量为：" + str(self.blood)
    def lose_blood(self,damage):
        self.blood -= damage
class Clip():
    def __init__(self,capacity):
        # 弹夹起始没有子弹
        self.capacity = capacity
        self.current_list = []
    def save_bullet(self,bullet):
        #当前子弹小于最大容量
        if len(self.current_list) < self.capacity:
            self.current_list.append(bullet)
    def __str__(self):
        return "当前弹夹的子弹数量为：" + str(len(self.current_list)) + "/" + str(self.capacity)
    def launch_bullet(self):
        # 判断当前弹夹中是否含有子弹
        if len(self.current_list) > 0:
            bullet = self.current_list[-1]
            # 如果有子弹就出弹
            self.current_list.pop()
            return bullet
        else:
            return None
class Bullet():
    def __init__(self,damage):
        #伤害力
        self.damage = damage
        #伤害敌人
    def hurt(self,enemy):
        enemy.lose_blood(self.damage)

# 创建一个战士
soldier = Person("许三多")
#创建一个弹夹
clip = Clip(20)
print(clip)
        
#添加五颗子弹
i = 0
while i < 5:
    #bullet = Bullet(4)
    #战士安装子弹到子弹夹
    soldier.install_bullet(clip, bullet)
    i += 1
#输出当前弹夹中的子弹数量
print(clip)        
class Gun():
    def __init__(self):
        #枪起始没有弹夹
        self.clip = None
    def __str__(self):
        if self.clip:
            return "枪当前有弹夹"
        else:
            return "枪没有弹夹"
    # 链接弹夹
    def mounting_clip(self,clip):
        if not self.clip:
            self.clip = clip
    # 射击
    def shoot(self,enemy):
        # 弹出子弹夹
        bullet = self.clip.launch_bullet()
        if bullet:
            bullet.hurt(enemy)
        else:
            print("没有子弹了放了空枪...")
#创建一个枪        
gun = Gun()
print(gun)
#安装弹夹
soldier.install_clip(gun,clip)
print(gun)
bullet = Bullet(5)

# 创建一个敌人
enemy = Person("敌人")
print(enemy)
#士兵拿枪
soldier.take_gun(gun)
#士兵开枪
soldier.fire(enemy)
print(clip)
print(enemy)
soldier.fire(enemy)
print(clip)
print(enemy)
```
