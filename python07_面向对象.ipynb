```python
# 设置对象属性

class Cat:
	def eat(self):
		#  哪一个对象调用的方法，self就是哪一个对象的应用
		print("%s爱吃鱼" % self.name)

	def drink(self):
		print("%s小猫要喝水"%self.name)

#  创建猫对象
tom = Cat()
tom.name = "Tom"
tom.eat()

tom.drink()
print(tom)


#  再创建一个猫对象
lazy_cat = Cat()
lazy_cat.name = "大懒猫"
lazy_cat.drink()
lazy_cat.eat()
print(lazy_cat)
```

```python
#__str__方法

class Cat:
	def __init__(self, new_name):
		self.name = new_name
		print("%s来了"%self.name)
	def __del__(self):
		print("%s我去了"%self.name)

	def __str__(self):
		# 必须要返回字符串
		return "我是小猫[%s]"%self.name
tom = Cat("Tom")
print(tom)
```

```python
# 案例演练 小明爱跑步

class Person:

	def __init__(self,name,weight):
		# self.属性 = 形参
		self.name = name
		self.weight = weight

	def __str__(self):

		return  "我的名字叫%s,体重是%.2f公斤"%(self.name,self.weight)
	def run(self):
		print("%s爱跑步，跑步锻炼身体"%self.name)
		self.weight -= 0.5
	def eat(self):
		print("%s爱吃东西，吃完这顿再减肥"%self.name)
		self.weight += 1

xiaoming = Person("小明", 75)
xiaoming.run()
xiaoming.eat()
print(xiaoming)

xiaomei = Person("小美", 40 )
xiaomei.eat()
xiaomei.run()
print(xiaomei)
print(xiaoming)
```

```python
# 摆放家具

class HouseItem:
	def __init__(self, name, item_area):
		self.name = name
		self.area = item_area
	def __str__(self):
		return ("%s 占地%.2f 平米"%(self.name,self.area))


class House:
	def __init__(self, house_type, house_area):
		self.type = house_type
		self.area = house_area
		self.free_area = house_area
		self.houseitem_list = []

	def __str__(self):

		return ("户型[%s]\n总面积[%.2f](剩余面积[%.2f])\n"
				"家具列表名称表%s\n"%
				(self.type, self.area, self.free_area,
				 self.houseitem_list))
	def add_item(self, item):
		print("要添加%s"%item)
		if item.area > self.free_area:
			print("剩余面积%.2f不够添加%s"%
				  (self.free_area,item))
			return

		self.houseitem_list.append(item.name)
		self.free_area -= item.area
bed = HouseItem("席梦思", 5)
chest = HouseItem("衣柜", 20)
table = HouseItem("餐桌", 1.5)

print(bed)
print(chest)
print(table)

house = House("二室一厅", 40)
house.add_item(bed)
house.add_item(chest)
house.add_item(table)

print(house)
```

```python
# 练习士兵突击

class Gun:
	# 初始化属性
	def __init__(self,name):
		self.name = name
		self.bullet_count = 0
	# 给枪装子弹
	def add_bullet(self,bullet_account):
		self.bullet_count += bullet_account
		print("加了[%d]个子弹"%bullet_account)
	# 让枪射击
	def shoot(self):
		# 先判断枪有没有子弹
		if self.bullet_count <= 0:
			print("[%s]没有子弹了！"%self.name)
			return
		# 射击完成后枪的子弹减少一
		self.bullet_count -= 1
		print("[%s] 突突突...[剩余子弹 %d]"%(self.name,self.bullet_count))



#  创建士兵
class Soldier:
	# 初始化枪的属性
	def __init__(self,name):
		self.name = name
		self.gun = None
	# 士兵进行射击调用枪对象
	def fire(self):
		#  判断有没有枪，没有枪则返回
		if self.gun is None:
			print("[%s]还没有枪"%self.name)
			return
		# 喊出响亮的口号
		print("冲啊！...[%s]"%(self.name))
		# 让枪装子弹
		self.gun.add_bullet(50)
		#  进行射击
		self.gun.shoot()

AK47 = Gun("AK47")
# AK47.add_bullet(50)
# AK47.shoot()

xusanduo = Soldier("许三多")
# 许三多拿了一把AK47枪
xusanduo.gun = AK47
# 进行射击了
xusanduo.fire()
print(xusanduo.gun)
```
